```python
import pandas as pd
import duckdb as db
```

```python
df_season_results = pd.read_csv('Data/march-machine-learning-mania-2025/MRegularSeasonDetailedResults.csv')
```

```python
df_season_results_W = df_season_results[['Season', 'WTeamID', 'WScore',
       'NumOT', 'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF']]
```

```python
df_season_results_W
```

```python
df_season_results_W['WLoc'].unique()
```

```python
db.query('''
with base 
as
(
    select Season, WTeamID, WLoc,
    case 
        when WLoc='N' then count(*) else 0 
    end as neutral_wins,
    case 
        when WLoc='A' then count(*) else 0 
    end as away_wins,
    case 
        when WLoc='H' then count(*) else 0 
    end as home_wins
    from df_season_results_W
    group by Season, WTeamID, WLoc
)
select Season, WTeamID, sum(neutral_wins) neutral_wins, sum(away_wins) away_wins, sum(home_wins) home_wins
from base
group by Season, WTeamID
''')
    
```
